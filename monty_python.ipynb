```python
import numpy as np 
import math
import random
import matplotlib.pyplot as plt
import logging
logger = logging.getLogger(__name__)
logging.basicConfig(
    filename="MONTY_PYTHON.log",
    encoding="utf-8",
    filemode="w",
    format="%(levelname)s:%(message)s",
    level=logging.DEBUG,
    # level=logging.WARNING,
)
from datetime import datetime
today=datetime.now().strftime('%Y-%m-%d %H:%M:%S')
logging.info('time: {}'.format(today))
rng = np.random.default_rng(19680801)

a = math.sqrt(math.log(4))
b = math.sqrt(2*math.pi)
s = a/(b-a)
inv_b = 1/b
ln_2 = math.log(2)
s_plus_one = s+1
s_times_b = s*b
e = 0.5718733*inv_b
g_de_x = inv_b/s
# Integrales
int_f_cero_a = 0.760968-(a*inv_b)
rotated_a_term = inv_b*(1-s)
log_rotated = math.log(rotated_a_term)
log_exp_xx = math.log(2/(math.sqrt(2*math.pi)))
# * math.log(math.pi)
variables = {'a': round(a,6), 'b': round(b,6), 's':round(s,7),
             'inv_b': round(inv_b,7),'ln_2': round(ln_2,7),
             's+1': round(s_plus_one,7),
             'e': round(e,6),
             'sb': round(s_times_b,6),
             'g(x)':round(g_de_x,6),
             'int_cero_a': round(int_f_cero_a,5),
             'rotated_a' : round(rotated_a_term,5),
             'log_rotated': round(log_rotated,5),
             'log_exp_xx': round(log_exp_xx,5)
             }

display(variables)
```

```python

# Definimos las variables z y w y las funciones para generar los nuevos valores
def rnor():
    while True:
        # Generamos x usando números aleatorios uniformes
        x = random.uniform(-b, b)  # Limites basados en la constante b
        logging.info('\n\n****** Generate x at x={}'.format(x))
        # Si el valor absoluto de x es menor que 1.17741, lo retornamos
        if abs(x) < 1.17741:
            logging.info('abs(x) = {}'.format(abs(x)))
            logging.info('x < a')
            logging.info(' if(1)============> x will be {}'.format(x))
            return x
        
        # Generamos el valor de y
        y = random.uniform(0,1)
        logging.info('****** Generate y at y={}'.format(y))
        
        # Comprobamos si x y y cumplen la condición, si es así, retornamos x
        logging.info("log(y) = {}, f(x) = {}".format(math.log(y),ln_2 - .5 * (x * x)))
        if math.log(y) < 0.6931472 - .5 * (x * x):
            logging.info('math.log(y) < .6931472 - .5 * (x * x)')
            logging.info(' if(2)============> x will be {}'.format(x))
            return x
        
        # Ajustamos el valor de x según su signo y el valor b = 2.506628, s = 0.8857913
        x = 0.8857913 * (2.506628 - x) if x > 0 else -0.8857913 * (2.506628 + x)
        logging.info('========= new_x={}'.format(x))
        # Verificamos si pasa esta otra condición (Tail)
        if math.log(1.8857931 - y) < 0.5718733 - 0.5 * (x * x):
            logging.info('math.log(1.8857931 - y) < .5718733 - 0.5 * (x * x)')
            logging.info(' if(3)============> x will be {}'.format(x))
            return x
        
        
        # Generamos otro número aleatorio v
        while True:
            v = random.uniform(-b,b)
            x = -math.log(math.fabs(v))*inv_b
            y = -math.log(random.uniform(0,1))
            
            # Si cumple la condición, rompemos el bucle y retornamos el valor final
            if y + y >= x * x:
                return 2.506628 + x if v>0 else -2.506628 - x
    
# Ejemplo de uso:
# random_value = rnor()
# logging.info('random_value={}'.format(random_value))
```
